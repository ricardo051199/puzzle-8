```python
import pandas as pd
```

```python
df = pd.read_csv('game1000.csv')
```

```python
df
```

```python
df['movimientos'] = df['movimientos'].apply(eval)
```

```python
# Obtener el mínimo y máximo de movimientos, y el algoritmo y heurística correspondientes
min_movimientos = df['movimientos'].apply(len).min()
max_movimientos = df['movimientos'].apply(len).max()
algoritmo_min_movimientos = df.loc[df['movimientos'].apply(len).idxmin(), 'algoritmo_busqueda']
heuristica_min_movimientos = df.loc[df['movimientos'].apply(len).idxmin(), 'heuristica']
algoritmo_max_movimientos = df.loc[df['movimientos'].apply(len).idxmax(), 'algoritmo_busqueda']
heuristica_max_movimientos = df.loc[df['movimientos'].apply(len).idxmax(), 'heuristica']
df['num_movimientos'] = df['movimientos'].apply(len)
promedio_movimientos = df['num_movimientos'].mean()
```

```python
# Calcular el mínimo, máximo y promedio del tiempo de resolución
min_tiempo_resolucion = df['tiempo_resolucion'].min()
max_tiempo_resolucion = df['tiempo_resolucion'].max()
algoritmo_min_resolucion = df.loc[df['tiempo_resolucion'].idxmin(), 'algoritmo_busqueda']
heuristica_min_resolucion = df.loc[df['tiempo_resolucion'].idxmin(), 'heuristica']
algoritmo_max_resolucion = df.loc[df['tiempo_resolucion'].idxmax(), 'algoritmo_busqueda']
heuristica_max_resolucion = df.loc[df['tiempo_resolucion'].idxmax(), 'heuristica']
promedio_tiempo_resolucion = df['tiempo_resolucion'].mean()
```

```python
# Obtener el mínimo, máximo y promedio del tamaño de frontera
min_tamaño_frontera = df['tamaño_frontera'].min()
max_tamaño_frontera = df['tamaño_frontera'].max()
algoritmo_min_tamaño_frontera = df.loc[df['tamaño_frontera'].idxmin(), 'algoritmo_busqueda']
heuristica_min_tamaño_frontera = df.loc[df['tamaño_frontera'].idxmin(), 'heuristica']
algoritmo_max_tamaño_frontera = df.loc[df['tamaño_frontera'].idxmax(), 'algoritmo_busqueda']
heuristica_max_tamaño_frontera = df.loc[df['tamaño_frontera'].idxmax(), 'heuristica']
promedio_tamaño_frontera = df['tamaño_frontera'].mean()
```

```python
# Contar cuántos `True` y `False` hay en la columna 'gano'
cuantos_gano = df['gano'].sum()
cuantos_no_gano = len(df) - cuantos_gano
```

```python
print("Mínimo de movimientos:", min_movimientos, algoritmo_min_movimientos, heuristica_min_movimientos)
print("Máximo de movimientos:", max_movimientos, algoritmo_max_movimientos, heuristica_max_movimientos)
print("Promedio movimientos:", promedio_movimientos)
print("Mínimo tiempo de resolución:", min_tiempo_resolucion, algoritmo_min_resolucion, heuristica_min_resolucion)
print("Máximo tiempo de resolución:", max_tiempo_resolucion, algoritmo_max_resolucion, heuristica_max_resolucion)
print("Promedio tiempo de resolución:", promedio_tiempo_resolucion)
print("Mínimo tamaño de frontera:", min_tamaño_frontera, algoritmo_min_tamaño_frontera, heuristica_min_tamaño_frontera)
print("Máximo tamaño de frontera:", max_tamaño_frontera, algoritmo_max_tamaño_frontera, heuristica_max_tamaño_frontera)
print("Promedio tamaño de frontera:", promedio_tamaño_frontera)
print("Cantidad de veces que se ganó:", cuantos_gano)
print("Cantidad de veces que no se ganó:", cuantos_no_gano)
```
